# Task 1. Vector lenghts

- Write a function that calculates vector lengths, answer should return array of shape $2$
- Apply this function to all the inputs

```python
def get_length(X):
    """Calculates the length (magnitude) of a vector X."""
    res = np.linalg.norm(X)  
    return res
```

```python
answers['lengths'] = [
    get_length(**inp) for inp in inputs
]
print("Answers (Lengths):", answers['lengths'])
```

# Task 2. Vector directions

- Write a function that calculates vector directions. Result should be of the same shape as the original array X
- Use $10^{-8}$ addition in division to avoid division by zero
- Apply this function to all arrays in the inputs

```python
def get_directions(X):
    """Calculates the direction (unit vector) of a vector X."""
    length = np.linalg.norm(X)
    if length == 0:
        return np.zeros_like(X) 
    res = X / (length + 10e-8)  
    return res
```

```python
answers['directions'] = [
    get_directions(**inp) for inp in inputs
]
print("\nAnswers (Directions):", answers['directions'])
```

# Task 4. Scalar Product

Write a function that calculates
- cosine distance between the vectors
- use $10^{-8}$ additional constant in the denominator for the sake of numerical stability

```python
def get_cosine(X):
    res = None
    v1 = X[:, 0]  # First vector
    v2 = X[:, 1]  # Second vector
    dot_product = np.dot(v1, v2)
    norm_v1 = np.linalg.norm(v1)
    norm_v2 = np.linalg.norm(v2)
    denominator = (norm_v1 * norm_v2) + 10e-8  # Add 10e-8 for stability
    res = dot_product / denominator
    return res
```

```python
answers['cosine'] = [
    get_cosine(**inp) for inp in inputs
]

print("Answers (Cosine):", answers['cosine'])
```

# Task 5. Orthonormalization

Create a pair of orthonormal vectors
- Take the first vector and normalize it
- Take the second vector, remove the collinear component to the first one
- Normalize it

```python
def get_orthonormal(X):
    res = X.copy().astype('float32')
    v1 = X[:, 0].astype('float32')  # First vector
    v2 = X[:, 1].astype('float32')  # Second vector

    # Normalize the first vector
    norm_v1 = np.linalg.norm(v1)
    if norm_v1 == 0:
        u1 = np.zeros_like(v1)
    else:
        u1 = v1 / norm_v1

    # Remove the collinear component of v2 to u1
    projection = np.dot(v2, u1) * u1
    v2_orthogonal = v2 - projection

    # Normalize the orthogonalized vector
    norm_v2_orthogonal = np.linalg.norm(v2_orthogonal)
    if norm_v2_orthogonal == 0:
        u2 = np.zeros_like(v2_orthogonal)
    else:
        u2 = v2_orthogonal / norm_v2_orthogonal

    res = np.column_stack((u1, u2))  # Stack the orthonormal vectors
    return res
```

```python
answers['orthonormal'] = [
    get_orthonormal(**inp) for inp in inputs
]
print("\nAnswers (Orthonormal):", answers['orthonormal'])
```
